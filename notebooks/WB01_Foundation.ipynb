## 1. Why This Paper Is the Analytical Anchor

Universal Health Coverage (UHC) is the centrepiece of United Nations Sustainable  
Development Goal 3.8. India, with 1.4 billion people and 36 states and union  
territories, presents one of the most complex UHC challenges in the world.  
National-level statistics mask enormous sub-national variation. State-level  
statistics, in turn, mask district-level variation.

Mukherji et al. (2024) is selected as the anchor paper for this capstone  
for four specific reasons:

**Reason 1 — It is the only validated district-level UHC index for India.**  
Prior work (Balarajan et al. 2011; Subramanian et al. 2006) measured health  
inequality at the state or socioeconomic-group level. Mukherji et al. are the  
first to construct a composite UHC index — the Composite Health District Index  
(CHDI) — validated against the WHO Service Coverage Index methodology,  
at the district level for all 707 NFHS-5 districts.

**Reason 2 — It uses the same data source as this capstone.**  
The CHDI is computed entirely from NFHS-5 (2019–21) district indicators —  
the same file loaded in this workbook. Every finding in this capstone is  
therefore directly comparable to Mukherji et al.'s published results.

**Reason 3 — It documents a finding that demands causal explanation.**  
Mukherji et al. report that 30.6% of total CHDI variance is *within* states,  
not between them. This means state-level governance and policy cannot explain  
the majority of the health coverage gap. District-level causal analysis is  
not merely useful — it is necessary.

**Reason 4 — The authors explicitly stop at description.**  
Mukherji et al. employ exclusively descriptive methods: concentration indices  
and inequality decomposition. The paper contains no causal model, no directed  
acyclic graph, no intervention analysis, and no fairness decomposition.  
These absences are the precise entry points for this capstone.

## 2. Five Research Gaps Identified in Mukherji et al. (2024)

A systematic reading of the paper reveals five distinct gaps. The first three  
are explicitly acknowledged by the authors. The fourth and fifth are identified  
through careful examination of what the framework structurally cannot do.

**Gap 1 — No causal inference.**  
The paper uses descriptive statistical methods: concentration indices and  
inequality decomposition. It identifies *where* disparities exist and  
*which variables correlate* with them. It cannot identify which variables  
*cause* them. There is no Directed Acyclic Graph, no do-calculus, and no  
Structural Causal Model. Correlation and causation are conflated throughout.  
*This capstone addresses Gap 1 through SCM-DAG construction and  
XGBoost-SHAP causal attribution.*

**Gap 2 — No equity weighting or fairness framework.**  
The CHDI assigns equal moral weight to all districts regardless of the  
origin of their deprivation. A district with low CHDI because of impassable  
terrain — a structural-physical constraint beyond any government's control —  
receives the same score as a district with low CHDI because of generations  
of caste-based exclusion from health services. Roemer (1998) established  
that these two types of disadvantage are morally distinct and require  
categorically different policy responses. Mukherji et al. make no such  
distinction.  
*This capstone addresses Gap 2 through the Tripartite Type I / II / III  
moral classification.*

**Gap 3 — No pathway attribution.**  
The composite index collapses infrastructure, empowerment, and insurance  
indicators into a single score. It is impossible to determine from the  
index alone whether a district's gap is driven primarily by lack of  
electricity (Type III, policy-actionable), child marriage prevalence  
(Type II, historically-produced), or mountainous terrain (Type I,  
structural-physical). Without pathway attribution, policy cannot be  
targeted.  
*This capstone addresses Gap 3 through SHAP-based pathway decomposition  
separating Path B (infrastructure) from Path C (empowerment).*

**Gap 4 — No intervention analysis or counterfactual reasoning.**  
A descriptive index cannot answer: what would the CHDI of district X be  
if PM-JAY reached full enrollment? What would it be if child marriage  
rates fell to zero? Answering these questions requires a causal model  
capable of do-calculus — computing P(CHDI | do(PM-JAY = 1)) rather than  
P(CHDI | PM-JAY = 1). Mukherji et al. work only at Rung 1 of Pearl's  
Causal Hierarchy (association). Policy simulation requires Rung 2  
(intervention) and Rung 3 (counterfactual).  
*This capstone addresses Gap 4 through G-computation and  
Callaway-Sant'Anna Difference-in-Differences identification.*

**Gap 5 — No cross-country transportability.**  
The framework is designed and validated exclusively for India. There is  
no test of whether the same causal structure — or the same Tripartite  
classification — holds in Nigeria, Bangladesh, Kenya, or Cambodia, all  
of which face structurally similar UHC challenges. Without a  
transportability test, the framework cannot claim generalisability  
and cannot inform global health policy.  
*This capstone addresses Gap 5 through application of the TCFD  
pipeline to 7 LMI countries in WB10-WB11.*

## 4. The CHDI Formula — What Mukherji et al. Computed

Having inspected the raw data in Cells 5 through 15, we now
understand exactly which columns are available, which are clean,
and which needed attention. We are now ready to map those columns
to the formula Mukherji et al. used to construct the CHDI.

### 4.1 Formula

Mukherji et al. constructed the CHDI as a geometric mean of
four domain sub-indices:

$$CHDI = \left( D_1 \times D_2 \times D_3 \times D_4 \right)^{1/4}$$

Each domain sub-index is itself a simple average of the
indicators assigned to that domain.

### 4.2 Domain Composition

| Domain | Label | Indicators from NFHS-5 |
|--------|-------|----------------------|
| Reproductive, Maternal, Newborn and Child Health | D1 | Family planning use, ANC ≥4 visits, Institutional births |
| Infectious Disease | D2 | Full vaccination, Stunting (inverted), Child anaemia (inverted) |
| Non-Communicable Disease | D3 | Women anaemia (inverted), Elevated blood pressure, High blood sugar |
| Service Capacity and Access | D4 | Health insurance coverage, Out-of-pocket expenditure (inverted) |

### 4.3 Inversion Rule

For indicators where a higher value means worse health
(stunting, anaemia, out-of-pocket expenditure), Mukherji et al.
apply the transformation:

$$Inverted = 100 - Observed$$

This ensures all indicators point in the same direction —
higher is always better — before the geometric mean is computed.

### 4.4 What We Will Do in Cells 17 and 18

- Cell 17 — identify the exact column from our raw file
  that maps to each indicator in the formula
- Cell 18 — compute CHDI for all 707 districts and compare
  our result to the published range in Mukherji et al. (2024)
  Table 2 (median 44.0%, range 18.8%–70.4%)

```python
# ============================================================
# CELL 17 — Map Raw Column Names to CHDI Domain Indicators
# ============================================================
# Purpose:
#   Document exactly which raw column feeds which domain.
#   Prints FOUND or NOT FOUND for every column.
#   A NOT FOUND here means Cell 18 will produce wrong results.
# ============================================================

# Domain 1 — Reproductive, Maternal, Newborn and Child Health
D1_cols = [
    "Current Use of Family Planning Methods (Currently Married Women Age 15-49  years) - Any method6 (%)",
    "Mothers who had at least 4 antenatal care visits  (for last birth in the 5 years before the survey) (%)",
    "Institutional births (in the 5 years before the survey) (%)"
]

# Domain 2 — Infectious Disease
# Stunting and child anaemia are inverted (100 - value)
D2_cols = [
    "Children age 12-23 months fully vaccinated based on information from either vaccination card or mother's recall11 (%)",
    "Children under 5 years who are stunted (height-for-age)18 (%)",
    "Children age 6-59 months who are anaemic (<11.0 g/dl)22 (%)"
]
D2_invert = [
    "Children under 5 years who are stunted (height-for-age)18 (%)",
    "Children age 6-59 months who are anaemic (<11.0 g/dl)22 (%)"
]

# Domain 3 — Non-Communicable Disease
# Women anaemia is inverted (100 - value)
D3_cols = [
    "Non-pregnant women age 15-49 years who are anaemic (<12.0 g/dl)22 (%)",
    "Women age 15 years and above wih Elevated blood pressure (Systolic ≥140 mm of Hg and/or Diastolic ≥90 mm of Hg) or taking medicine to control blood pressure (%)",
    "Women age 15 years and above wih high or very high (>140 mg/dl) Blood sugar level or taking medicine to control blood sugar level23 (%)"
]
D3_invert = [
    "Non-pregnant women age 15-49 years who are anaemic (<12.0 g/dl)22 (%)"
]

# Domain 4 — Service Capacity and Access
# OOP expenditure is inverted
D4_cols = [
    "Households with any usual member covered under a health insurance/financing scheme (%)",
    "Average out-of-pocket expenditure per delivery in a public health facility (for last birth in the 5 years before the survey) (Rs.)"
]
D4_invert = [
    "Average out-of-pocket expenditure per delivery in a public health facility (for last birth in the 5 years before the survey) (Rs.)"
]

# Print mapping with FOUND / NOT FOUND
all_domains = [
    ("D1 — Reproductive, Maternal, Newborn and Child Health", D1_cols, []),
    ("D2 — Infectious Disease",                               D2_cols, D2_invert),
    ("D3 — Non-Communicable Disease",                         D3_cols, D3_invert),
    ("D4 — Service Capacity and Access",                      D4_cols, D4_invert),
]

for domain_name, cols, inverted in all_domains:
    print(f"{domain_name}")
    for col in cols:
        inv_flag = "  [INVERTED]" if col in inverted else ""
        present  = "FOUND" if col in df_sheet1.columns else "NOT FOUND"
        print(f"  [{present}]{inv_flag}")
        print(f"    {col}")
    print()
```

```python
# ============================================================
# CELL 18 — Compute CHDI for All 707 Districts
# ============================================================
# Purpose:
#   Apply the Mukherji et al. formula to our raw data.
#   Print every district's D1, D2, D3, D4, and CHDI.
#   Compare summary statistics to the published values.
# ============================================================

# Step 1 — make a working copy of the data
df = df_sheet1.copy()

# Step 2 — convert all domain columns to numeric
all_domain_cols = D1_cols + D2_cols + D3_cols + D4_cols
for col in all_domain_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Step 3 — apply inversion where required (100 - value)
all_inverted = D2_invert + D3_invert + D4_invert
for col in all_inverted:
    df[col] = 100 - df[col]

# Step 4 — compute each domain sub-index as simple mean
df["D1"] = df[D1_cols].mean(axis=1)
df["D2"] = df[D2_cols].mean(axis=1)
df["D3"] = df[D3_cols].mean(axis=1)
df["D4"] = df[D4_cols].mean(axis=1)

# Step 5 — compute CHDI as geometric mean of four domains
df["CHDI"] = (df["D1"] * df["D2"] * df["D3"] * df["D4"]) ** (1/4)

# Step 6 — print results for every district
h1,h2,h3,h4,h5,h6,h7 = "District","State","D1","D2","D3","D4","CHDI"
print(f"{h1:<35} {h2:<30} {h3:>6} {h4:>6} {h5:>6} {h6:>6} {h7:>6}")
print("-" * 100)

for _, row in df.iterrows():
    print(f"  {str(row['District Names']):<33} "
          f"{str(row['State/UT']):<28} "
          f"{row['D1']:>6.1f} "
          f"{row['D2']:>6.1f} "
          f"{row['D3']:>6.1f} "
          f"{row['D4']:>6.1f} "
          f"{row['CHDI']:>6.1f}")

# Step 7 — print summary statistics
print()
print("Summary statistics:")
print(f"  Districts computed : {df['CHDI'].notna().sum()}")
print(f"  Median CHDI        : {df['CHDI'].median():.2f}%")
print(f"  Min CHDI           : {df['CHDI'].min():.2f}%"
      f"  — {df.loc[df['CHDI'].idxmin(), 'District Names']}")
print(f"  Max CHDI           : {df['CHDI'].max():.2f}%"
      f"  — {df.loc[df['CHDI'].idxmax(), 'District Names']}")
print()
print("Mukherji et al. (2024) published values:")
print(f"  Median : 44.0%")
print(f"  Min    : 18.8%  — Nandurbar, Maharashtra")
print(f"  Max    : 70.4%  — Nilgiris, Tamil Nadu")
```

```python
# ============================================================
# CELL 19 — Plot CHDI Distribution
# ============================================================
# Purpose:
#   Visualise the CHDI distribution across all 707 districts.
#   Shows our median vs Mukherji median vs SDG 50% target.
# ============================================================

fig, ax = plt.subplots(figsize=(12, 6))

ax.hist(
    df["CHDI"].dropna(),
    bins=40,
    color="#1B6B8A",
    alpha=0.7,
    edgecolor="white",
    linewidth=0.5
)

df["CHDI"].dropna().plot.kde(
    ax=ax,
    color="#0C2340",
    linewidth=2
)

ax.axvline(
    df["CHDI"].median(),
    color="#C0392B",
    linewidth=2,
    linestyle="--",
    label=f"Our median: {df['CHDI'].median():.1f}%"
)

ax.axvline(
    44.0,
    color="#D4833A",
    linewidth=2,
    linestyle=":",
    label="Mukherji median: 44.0%"
)

ax.axvline(
    50.0,
    color="#10653E",
    linewidth=2,
    linestyle="-.",
    label="SDG 3.8 target: 50%"
)

ax.set_title(
    "CHDI Distribution Across 707 Districts — NFHS-5 (2019-21)",
    fontsize=14,
    fontweight="bold",
    pad=15
)
ax.set_xlabel("CHDI (%)", fontsize=12)
ax.set_ylabel("Number of Districts", fontsize=12)
ax.legend(fontsize=10)

plt.tight_layout()
plt.savefig("WB01_Fig1_CHDI_Distribution.png", dpi=150)
plt.show()

print(f"Districts below 50% (SDG target) : "
      f"{(df['CHDI'] < 50).sum()} "
      f"({(df['CHDI'] < 50).mean()*100:.1f}%)")
print(f"Districts above 50%              : "
      f"{(df['CHDI'] >= 50).sum()} "
      f"({(df['CHDI'] >= 50).mean()*100:.1f}%)")
```

```python
# ============================================================
# CELL 20 — Within-State vs Between-State Variance Decomposition
# ============================================================
# Purpose:
#   Replicate the key finding from Mukherji et al. (2024):
#   30.6% of CHDI variance is within states, not between them.
#   This validates our computation and justifies why
#   district-level analysis is necessary.
# ============================================================

# Step 1 — compute grand mean and state means
grand_mean   = df["CHDI"].mean()
state_means  = df.groupby("State/UT")["CHDI"].mean()
state_counts = df.groupby("State/UT")["CHDI"].count()

# Step 2 — compute total sum of squares
SST = ((df["CHDI"] - grand_mean) ** 2).sum()

# Step 3 — compute between-state sum of squares
SSB = (state_counts * (state_means - grand_mean) ** 2).sum()

# Step 4 — compute within-state sum of squares
SSW = SST - SSB

# Step 5 — compute percentages
pct_between = (SSB / SST) * 100
pct_within  = (SSW / SST) * 100

# Step 6 — print results
print("Variance Decomposition Results:")
print()
print(f"  Grand mean CHDI          : {grand_mean:.2f}%")
print(f"  Total sum of squares     : {SST:.2f}")
print(f"  Between-state SS         : {SSB:.2f}")
print(f"  Within-state SS          : {SSW:.2f}")
print()
print(f"  Between-state variance   : {pct_between:.1f}%")
print(f"  Within-state variance    : {pct_within:.1f}%")
print()
print("Mukherji et al. (2024) published values:")
print(f"  Between-state variance   : 69.4%")
print(f"  Within-state variance    : 30.6%")
print()

# Step 7 — print state level detail
h1,h2,h3,h4 = "State/UT","Districts","Mean CHDI","Std CHDI"
print(f"{h1:<40} {h2:>10} {h3:>10} {h4:>10}")
print("-" * 73)

state_std = df.groupby("State/UT")["CHDI"].std()

for state in sorted(state_means.index):
    print(f"  {state:<38} "
          f"{state_counts[state]:>10} "
          f"{state_means[state]:>10.2f} "
          f"{state_std[state]:>10.2f}")

print()

# Step 8 — validation check
diff_within = abs(pct_within - 30.6)
if diff_within < 2.0:
    print("VALIDATION PASSED — within-state variance matches "
          "Mukherji et al. within 2 percentage points.")
else:
    print(f"VALIDATION NOTE — within-state variance differs "
          f"by {diff_within:.1f}pp from Mukherji et al.")
    print("Review domain column mappings in Cell 17.")
```
